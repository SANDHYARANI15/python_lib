Font properties of label&title:
fontdict->defines of dictionary file is
f={'family':'arial',size:20,'color':'r'} these values are passes to fontdic)


```python
x=np.arange(0,15,3)
y=[i*2 for i in x ]
f={'family':'arial','color':'y','size':17}
plt.title("Simple graph",fontdict=f)
plt.xlabel("Data for x",fontdict=f)
plt.ylabel("Data for y",fontdict=f)
plt.plot(x,y,marker="o")
```

```python
x=np.arange(0,15,3)
y=[i*2 for i in x ]
f={'family':'arial','color':'y','size':17}
plt.title("Simple graph",fontdict=f)
plt.xlabel("Data for x",fontdict=f)
plt.ylabel("Data for y",fontdict=f)
plt.grid(axis="both",lw=1,ls="dotted",color="r")
plt.plot(x,y,marker="o")
```

subplots:
    in the figuare can have multi graph
3 argguments have
subplot(r,c,i)
R->row
c->colomn
I->index&plot
suptitle->mojar title on all subgraph


```python
x=np.arange(0,20,3)
y=[i*3 for i in x ]
y1=[i*2 for i in x ]
plt.suptitle("Multi graph")
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("plot-1")
plt.subplot(1,2,2)
plt.plot(x,y1)
plt.title("plot-2")
plt.show()
```

```python
#without parameters
x=np.arange(0,15,3)
y=[i*2 for i in x ]
y1=[i*2 for i in x ]
f={'family':'arial','color':'y','size':17}
plt.title("Simple graph",fontdict=f)
plt.xlabel("Data for x",fontdict=f)
plt.ylabel("Data for y",fontdict=f)
plt.plot(x,y,y1,marker="o",label="SQUARE")
plt.legend(['CUBES','SQUARE'],loc="upper right",framealpha=1,facecolor="y",edgecolor="r",fancybox=True,shadow=True,fontsize=10)
plt.show()
```

grouped or multiple graph :



```python
years=["2017","2018","2019","2020"]
cse=[74,82,89,79]
it=[58,72,76,61]
ece=[67,74,81,68]
eee=[64,21,43,54]
w=0.2
cse_bar=np.arange(len(years))
plt.bar(cse_bar,cse, width=w,label="cse")
it_bar=[i+w for i in cse_bar]
ece_bar=[i+w for i in it_bar]
plt.bar(it_bar,it, width=w,label="it")
plt.bar(ece_bar,it, width=w,label="ece")
eee_bar=[i+w for i in ece_bar]
plt.bar(eee_bar,it, width=w,label="eee")
plt.xticks(cse_bar+w+w/2,years)
plt.
plt.title("placemecrts")
plt.legend()
plt.show()
```
